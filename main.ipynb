```python
api_key = "API_KEY"

from openai import OpenAI

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=api_key,
)



with open("inta_texts_cleaned.txt", "r", encoding="utf-8") as f:
    text = f.read()

print("سلام به دستیار هوشمند امور مالیاتی خوش آمدیدید. سوال خود را بپرسید و من آماده پاسخگویی هستم")

while True:
    
    question = input("You: ")
    if question.lower() == "exit":
        print("Goodbye!")
        break

    print("You:", question)

    messages = [
        {"role": "system", "content": "You are a helpful assistant. Answer based on the provided context."},
        {"role": "user", "content": f"Context:\n{text}\n\nQuestion:\n{question}"}
    ]



    completion = client.chat.completions.create(
            model="openai/gpt-oss-120b:free",
            messages=messages
        )


    print("Bot:", completion.choices[0].message.content)
```
